## Relazione tra durata di una telefonata e il suo costo

La gradezza indipendente è il tempo $t$ espresso in secondi.

La grandezza dipendente è costo $C$ espresso in €.

$0.10$ € costo al minuto

$0.15$ € scatto alla risposta

La relazione tra $t$ e $C$ è data da:
$$
C = 0.15 + \frac{0.10t}{60}
$$

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
durata = 600
t = np.linspace(0, durata, 100)
```

```python
C1 = 0.15 + (0.10*t)/60

C2 = 0.20 + (0.05*t)/60
```

```python
plt.plot(t, C1, label='Tariffa A')
plt.plot(t, C2, label='Tariffa B')
plt.legend()
plt.grid()
plt.title('Costo in funzione della durata')
plt.xlabel('Durata [s]')
plt.ylabel('Costo [€]')
plt.show()
```
