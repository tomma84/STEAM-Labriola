```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
densita =   [7.8,       2.708,        8.96]
materiali = ["Acciaio", "Alluminio", "Rame"]
```

```python
volumi = np.array( [7.5,   3 ] )
masse = np.array(  [64.07, 15.08 ] )
```

```python
x = np.array( [0, 10] )

y =densita[0] * x
plt.plot(x, y, label=materiali[0])

y =densita[1] * x
plt.plot(x, y, label=materiali[1])

y =densita[2] * x
plt.plot(x, y, label=materiali[2])

plt.plot(volumi, masse, '.r')

plt.xlabel('Volume')
plt.ylabel('Massa')

plt.legend()
plt.show()
```
