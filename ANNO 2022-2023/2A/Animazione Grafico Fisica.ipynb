```python
import numpy as np
import matplotlib.pyplot as plt

from matplotlib.animation import FuncAnimation

%matplotlib notebook
%config InlineBackend.figure_format = 'retina'
```

```python
durata = 20
fps = 30
frame_totali = durata*fps
delta_t = 1/fps * 1000

t = np.arange(0, durata, 0.1)

# LEGGI ORARIE DEI MOTI
posizioneA = -16 + 9*t
posizioneB = 8*t
```

```python
fig = plt.figure()

plt.plot(t, posizioneA, 'b--', label='Moto primo corridore')
plt.plot(t, posizioneB, 'r--', label='Moto secondo corridore')

plt.plot(16, 128, 'xg', label='Punto di incontro')

plt.plot(0, -16, 'xb', label="Posizione iniziale di A")
plt.plot(0, 0, 'xr', label="Posizione iniziale di B")

motoA = plt.plot([], 'bo')[0]
motoB = plt.plot([], 'ro')[0]

def animate(frame):
    
    t = frame/frame_totali*durata
    
    sa = -16 + 9*t
    sb = 8*t
    
    motoA.set_data(t,sa)
    motoB.set_data(t,sb)


plt.xlabel('Tempo (s)')
plt.ylabel('Posizione (m)')

plt.xlim(-0.5, 20.5)
plt.ylim(-25, 175)

plt.legend()

plt.grid()

anim = FuncAnimation(fig, animate, frames=frame_totali, interval=delta_t)
plt.show()
```
