```python
import matplotlib.pyplot as plt
import numpy as np

from matplotlib.animation import FuncAnimation
from matplotlib.animation import FFMpegWriter

%matplotlib notebook
%config InlineBackend.figure_format = 'retina'
```

```python
fps = 30
durata = 20 
max_frame = fps*durata

fig = plt.figure()

motoA = plt.plot([], 'b')[0]
motoB = plt.plot([], 'r')[0]

plt.xlim(-0.5, 20.5)
plt.ylim(-25, 175)
plt.grid()

def animate(frame):
    t = np.arange(0, frame/max_frame*durata, 0.1)
    sa = -16 + 9*t
    sb = 8*t
    motoA.set_data((t,sa))
    motoB.set_data((t,sb))

anim = FuncAnimation(fig, animate, frames=max_frame, interval=1/fps*1000)
plt.show()
```

```python
writervideo = FFMpegWriter(fps=fps) 
anim.save("video.mp4", writer=writervideo)
```

```python
fps = 30
durata = 20 
max_frame = fps*durata

fig = plt.figure()

t = np.arange(0, durata, 0.1)
sa = -16 + 9*t
sb = 8*t

plt.plot(t,sa, 'b--')
plt.plot(t,sb, 'r--')

motoA = plt.plot([], 'ob', label='Primo altleta')[0]
motoB = plt.plot([], 'or', label='Secondo altleta')[0]

plt.plot(0,-16, 'xb', label='p(t=0) - Primo atleta')
plt.plot(0,0, 'xr', label='p(t=0) - Secondo atleta')

plt.plot(16, 128, 'xg', label='Punto di incontro')

plt.xlabel('Tempo (s)')
plt.ylabel('Spazio (m)')

plt.legend(loc='upper left')

plt.xlim(-0.5, 20.5)
plt.ylim(-25, 175)
plt.grid()

def animate(frame):
    t = frame/max_frame*durata
    sa = -16 + 9*t
    sb = 8*t
    motoA.set_data((t,sa))
    motoB.set_data((t,sb))

anim = FuncAnimation(fig, animate, frames=max_frame, interval=1/fps*1000)
plt.show()
```
