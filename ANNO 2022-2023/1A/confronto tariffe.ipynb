# Confronto tariffe

Confrontiamo due possibili tariffe telefoniche

| Scatto alla risposta | Costo al minuto |
| -------------------- | --------------- |
|                 0.15 |            0.10 |
|                 0.20 |            0.05 |

Il legame tra il costo $c$ e la durata della telefonta $t$ è dato dalla relazione:
$$
    c = R +C t
$$
dove:
- $R$ rappresenta lo scatto alla risposta
- $C$ costo al minuto

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 5, 31)
c1 = 0.15 + 0.10*t
c2 = 0.20 + 0.05*t
```

```python
plt.plot(t, c1, 'b', label='Prima tariffa')
plt.plot(t, c2, 'r', label='Seconda tariffa')

plt.legend()
plt.grid()
plt.title('Confronto tra tariffe telefoniche')
plt.xlabel('Tempo (m)')
plt.ylabel('Costo (€)')
plt.show()
```
