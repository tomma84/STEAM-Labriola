```python
import matplotlib.pyplot as plt
```

```python
def simulazione(pop_iniziale, durata, r):
    
    x = pop_iniziale
    lista_x = [x]
    
    for i in range(durata):
        x = r*x*(1-x)
        lista_x.append(x)
        
    plt.plot(lista_x)
    plt.title(f'Simulazione con r = {r}')
    plt.grid()
    plt.show()
```

```python
simulazione(0.40, 100, 1.2)
```

```python
simulazione(1, 100, 1.2)
```
