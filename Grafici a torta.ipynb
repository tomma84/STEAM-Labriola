```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
dati =      [12,       23,          45,       10]
etichette = ['Calcio', 'Pallavolo', 'Tennis', 'Golf']
explode =   [0,        0.2,           0,       0]
```

```python
plt.pie(dati, labels=etichette, autopct='%.1f%%', explode=explode, shadow=True, startangle=90)
plt.savefig('grafico.png', dpi=100)
plt.show()
```
