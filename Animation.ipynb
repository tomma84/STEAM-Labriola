```python
fig = plt.figure()

lines = plt.plot([])
line1 = lines[0]
lines = plt.plot([])
line2 = lines[0]


plt.xlim(0, 100)
plt.ylim(0, 300)


def animate(frame):
    x = np.arange(0, frame, 0.1)
    y1 = 3*x
    y2 = 2*x+100
    line1.set_data((x,y1))
    line2.set_data((x,y2))
    

anim = FuncAnimation(fig, animate, frames=100, interval=5)
plt.show()        
```
