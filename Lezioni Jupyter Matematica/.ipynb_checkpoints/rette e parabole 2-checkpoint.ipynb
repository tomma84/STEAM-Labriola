# Equazioni di primo e secondo grado

Rette e prabole corrispondo alle rappresentazioni grafiche di equazioni di primo e secondo grado.

Esempio di equazione di una retta:
$$
y = 3x + 4
$$

Esempio di equazione di una parabola:
$$
y = -x^2 + x -1
$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5, 5, 100)
```

```python
len(x)
```

```python
y1 = 3*x + 4

px = 0
py = 4
```

```python
plt.title('Grafico di un retta')
plt.plot(x, y1, label='Retta')
plt.plot(px, py, 'or', label='P(0,q)')

plt.legend()

plt.xticks([0])
plt.yticks([0, 4])

plt.grid()
plt.show()
```

```python
y2 = -x**2 +x -1
```

```python
plt.title('Grafico di una parabola')
plt.plot(x, y2, label='Parabola')

plt.legend()

plt.grid()
plt.show()
```

```python
plt.title('Grafici di retta e parabola')

plt.plot(x, y1, label='Retta')
plt.plot(x, y2, label='Parabola')

plt.legend()

plt.grid()
plt.show()
```

## Secondo esempio di parabola

$$
y = 2x^2 -4x
$$

```python
x = np.linspace(-2, 4, 100)

a = 2
b = -4
c = 0

y3 = a*x**2 +b*x +c

x_asse_simmetria = [  1,  1]
y_asse_simmetria = [-5, 20]

x_int_1 = 0
y_int_1 = 0

x_int_2 = 2
y_int_2 = 0
```

```python
plt.plot(x, y3, label='Parabola')
plt.plot(x_asse_simmetria, y_asse_simmetria, label='Asse simmetria')
plt.plot(x_int_1, y_int_1, '.r')
plt.plot(x_int_2, y_int_2, '.r')

plt.grid()
plt.legend()
plt.show()
```

# Intersezione tra parabola e rette

$$
\begin{cases}
    y=0 \\
    y = 2x^2 -4x \\
\end{cases}
$$

$$
0 = 2x^2 -4x
$$

```python
import sympy
```

```python
x = sympy.Symbol('x')
```

```python
primo_membro = 0
secondo_membro = 2*x**2 - 4*x
```

```python
equazione = sympy.Eq(primo_membro, secondo_membro)
```

```python
equazione
```

```python
sympy.solve(equazione, x)
```

# Funzione positiva e negativa

```python
x1 = np.linspace(-2, 0)
x2 = np.linspace(0, 2)
x3 = np.linspace(2, 4)
```

```python
y1 = 2*x1**2 -4*x1
y2 = 2*x2**2 -4*x2
y3 = 2*x3**2 -4*x3
```

```python
plt.plot(x1, y1, 'b')
plt.plot(x2, y2, 'r')
plt.plot(x3, y3, 'b')
plt.grid()
plt.show()
```
