# Rette e Parabole

Le rette e le parabole sono rappresentazioni grafiche di alcune funzioni.

Esempio di equazione di primo grado:

$$
y = x + 1 
$$

Esempio di equazione di secondo grado:

$$
y = 3x^2 +x -2
$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-2, 2, 100)
```

```python
y1 = x + 1
```

```python
plt.title('Funzione di primo grado')
plt.plot(x, y1)
plt.grid()
plt.show()
```

```python
y2 = 3*x**2 +x -2
px = -1
py = 0
```

```python
plt.title('Funzione di secondo grado')
plt.plot(x, y2)
plt.grid()
plt.show()
```

```python
plt.title('Funzioni di primo e secondo grado')

plt.plot(x, y1, label='Retta')
plt.plot(x, y2, label='Parabola')
plt.plot(px, py, '.r', label='A')

plt.grid()

plt.legend()

plt.xticks([-1, 0]) # ASSE Y
plt.yticks([0]) # ASSE X

plt.show()
```

```python
x = np.linspace(-4, 2, 100)
y3 = 2*x**2 + 4*x

x_asse = [-1, -1]
y_asse = [-3, 16]
```

```python
plt.title('Funzione di secondo grado')
plt.plot(x, y3)
plt.plot(x_asse, y_asse)
plt.grid()
plt.show()
```

# Soluzione Equazione

```python
import sympy
```

```python
x = sympy.Symbol('x')
```

```python
primo_membro = x + 1
secondo_membro = 3*x**2 +x -2
```

```python
equazione = sympy.Eq(primo_membro, secondo_membro)
```

```python
sympy.solve(equazione, x)
```
