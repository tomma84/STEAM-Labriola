```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dati = pd.read_csv('dati2.csv')
```

```python
dati
```

```python
plt.plot(dati.time,dati.TgF)
plt.show()
```

```python
fig, axs = plt.subplots(3, 1,figsize=(10,10))

fig.suptitle('Accelerometro')

axs[0].plot(dati.time, dati.gFx, 'r')
axs[0].title.set_text('Asse X')

axs[1].plot(dati.time, dati.gFy, 'g')
axs[1].title.set_text('Asse Y')

axs[2].plot(dati.time, dati.gFz, 'b')
axs[2].title.set_text('Asse Z')


for grafico in axs:
    grafico.grid()

fig.savefig('Accelerometro.jpg')
```
